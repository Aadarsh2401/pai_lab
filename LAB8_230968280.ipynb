```python
class SmartLightingAgent:
    def __init__(self):
        self.KB = {}

    def perceive(self, motion, dark, override):
        self.KB["MotionDetected"] = motion
        self.KB["IsDark"] = dark
        self.KB["ManualOverride"] = override

    def infer(self):
        if self.KB["ManualOverride"]:
            return "Manual Control - No Action"

        if self.KB["IsDark"] and self.KB["MotionDetected"]:
            return "Light ON"

        if not self.KB["MotionDetected"]:
            return "Light OFF"

        return "No Change"

agent = SmartLightingAgent()

test_cases = [
    (True, True, False),   # Motion + Dark → ON
    (False, True, False),  # No Motion → OFF
    (True, False, False),  # Motion but not dark → No Change
    (False, False, False), # No motion, not dark → OFF
    (True, True, True),    # Override → Ignore
]

for i, (m, d, o) in enumerate(test_cases, 1):
    agent.perceive(m, d, o)
    print(f"Test {i}: {agent.infer()}")
```

```python
class MedicalAgent:
    def __init__(self):
        self.KB = set()
        self.inferred = set()

    def add_symptoms(self, fever, cough, chest_pain):
        if fever:
            self.KB.add("Fever")
        if cough:
            self.KB.add("Cough")
        if chest_pain:
            self.KB.add("ChestPain")

    def forward_chain(self):
        changed = True
        while changed:
            changed = False

            if {"Fever", "Cough"}.issubset(self.KB) and "Flu" not in self.KB:
                self.KB.add("Flu")
                changed = True

            if {"Fever", "Cough", "ChestPain"}.issubset(self.KB) and "Pneumonia" not in self.KB:
                self.KB.add("Pneumonia")
                changed = True

            if "Pneumonia" in self.KB and "HospitalizationRequired" not in self.KB:
                self.KB.add("HospitalizationRequired")
                changed = True

    def get_results(self):
        return self.KB

cases = [
    (True, True, False),   # Flu only
    (True, True, True),    # Pneumonia + Hospitalization
    (False, True, True),   # No disease
    (True, False, False),  # No disease
    (False, False, False)  # Nothing
]

for i, (f, c, cp) in enumerate(cases, 1):
    agent = MedicalAgent()
    agent.add_symptoms(f, c, cp)
    agent.forward_chain()
    print(f"Test {i}: {agent.get_results()}")
```

```python
class TrafficAgent:
    def __init__(self):
        self.KB = {}

    def perceive(self, speed, red, crossed, emergency):
        self.KB = {
            "SpeedAboveLimit": speed,
            "SignalRed": red,
            "VehicleCrossedLine": crossed,
            "EmergencyVehicle": emergency
        }

    def infer(self):
        if self.KB["EmergencyVehicle"]:
            return "No Violation (Emergency)"

        violations = []

        if self.KB["SpeedAboveLimit"]:
            violations.append("SpeedViolation")

        if self.KB["SignalRed"] and self.KB["VehicleCrossedLine"]:
            violations.append("SignalViolation")

        return violations if violations else ["No Violation"]

agent = TrafficAgent()

timeline = [
    (True, False, False, False),   # Speed only
    (False, True, True, False),    # Signal violation
    (True, True, True, False),     # Both violations
    (True, True, True, True),      # Emergency override
    (False, False, False, False)   # Clean
]

for t, case in enumerate(timeline, 1):
    agent.perceive(*case)
    print(f"Time {t}: {agent.infer()}")
```

```python
class CyberSecurityAgent:
    def __init__(self):
        self.KB = set()

    def perceive(self, failed_logins, unknown_ip, sensitive_access, admin):
        if failed_logins:
            self.KB.add("MultipleFailedLogins")
        if unknown_ip:
            self.KB.add("LoginFromUnknownIP")
        if sensitive_access:
            self.KB.add("AccessToSensitiveFiles")
        if admin:
            self.KB.add("AdminPrivileges")

    def infer(self):
        if {"MultipleFailedLogins", "LoginFromUnknownIP"}.issubset(self.KB):
            self.KB.add("SuspiciousLogin")

        if "SuspiciousLogin" in self.KB and "AdminPrivileges" in self.KB:
            self.KB.add("HighRiskIntrusion")

        if "HighRiskIntrusion" in self.KB:
            return "ALERT: Intrusion Detected"

        return "System Normal"

cases = [
    (True, True, False, False),   # Suspicious login
    (True, True, False, True),    # High risk → Alert
    (False, True, True, True),    # No suspicious login
    (True, False, False, True),   # No suspicious login
    (False, False, False, False)  # Normal
]

for i, case in enumerate(cases, 1):
    agent = CyberSecurityAgent()
    agent.perceive(*case)
    print(f"Test {i}: {agent.infer()}")
```

```python
class WarehouseRobot:
    def __init__(self):
        self.KB = {}

    def perceive(self, obstacle, path_clear, fragile, human, reached):
        self.KB = {
            "ObstacleAhead": obstacle,
            "PathClear": path_clear,
            "CarryingFragileItem": fragile,
            "HumanNearby": human,
            "DestinationReached": reached
        }

    def infer(self):
        if self.KB["DestinationReached"]:
            return "Stop (Reached)"

        if self.KB["CarryingFragileItem"] and self.KB["ObstacleAhead"]:
            return "SlowDown"

        if self.KB["ObstacleAhead"]:
            return "Stop"

        if self.KB["HumanNearby"]:
            return "ReduceSpeed"

        if self.KB["PathClear"] and not self.KB["HumanNearby"]:
            return "MoveForward"

        return "Idle"

robot = WarehouseRobot()

cases = [
    (True, False, False, False, False),   # Obstacle → Stop
    (True, False, True, False, False),    # Fragile + Obstacle → SlowDown
    (False, True, False, True, False),    # Human → ReduceSpeed
    (False, True, False, False, False),   # Clear path → MoveForward
    (False, True, False, False, True),    # Destination reached
]

for i, case in enumerate(cases, 1):
    robot.perceive(*case)
    print(f"Test {i}: {robot.infer()}")
```
