```python
import random

class SimpleReflexVacuumAgent:
    def act(self, percept):
        x, y, status = percept

        # Rule 1: If current room is dirty, clean it
        if status == 'Dirty':
            return 'Suck'

        # Rule 2: If clean, move randomly (valid moves only)
        moves = []

        if x > 0:
            moves.append('Up')
        if x < 1:
            moves.append('Down')
        if y > 0:
            moves.append('Left')
        if y < 1:
            moves.append('Right')

        return random.choice(moves)
```

```python
# Create 2x2 environment with random dirt
def create_environment():
    env = {}
    for x in range(2):
        for y in range(2):
            env[(x, y)] = random.choice(['Clean', 'Dirty'])
    return env


# Run simulation
env = create_environment()
agent = SimpleReflexVacuumAgent()

# Initial agent position
agent_pos = (0, 0)

steps = 10
moves = 0
cleaned = 0

for step in range(steps):
    x, y = agent_pos
    percept = (x, y, env[(x, y)])

    action = agent.act(percept)

    print(f"Step {step+1} | Percept: {percept} | Action: {action}")

    if action == 'Suck':
        if env[(x, y)] == 'Dirty':
            cleaned += 1
        env[(x, y)] = 'Clean'

    else:
        moves += 1
        if action == 'Up':
            agent_pos = (x - 1, y)
        elif action == 'Down':
            agent_pos = (x + 1, y)
        elif action == 'Left':
            agent_pos = (x, y - 1)
        elif action == 'Right':
            agent_pos = (x, y + 1)

print("\nSimulation Summary")
print("Rooms cleaned:", cleaned)
print("Total movements:", moves)
```

```python
class SimpleModelVacuumAgent:
    def __init__(self):
        self.model = {'A': None, 'B': None}
        self.location = 'A'

    def act(self, percept):
        location, status = percept

        self.model[location] = status

        if status == 'Dirty':
            return 'Suck'

        if location == 'A' and self.model['B'] != 'Clean':
            self.location = 'B'
            return 'Right'

        if location == 'B' and self.model['A'] != 'Clean':
            self.location = 'A'
            return 'Left'

        return 'NoOp'
```

```python
# Environment setup
environment = {
    'A': 'Dirty',
    'B': 'Dirty'
}

# Create agent
agent = SimpleModelVacuumAgent()

# Statistics
steps = 6
cleaned = 0
movements = 0

print("Initial Environment:", environment)
print("Initial Model:", agent.model)
print("-" * 40)

for step in range(steps):
    location = agent.location
    percept = (location, environment[location])

    action = agent.act(percept)

    print(f"Step {step+1}")
    print("Percept:", percept)
    print("Action:", action)

    if action == 'Suck':
        if environment[location] == 'Dirty':
            cleaned += 1
        environment[location] = 'Clean'

    elif action in ['Left', 'Right']:
        movements += 1

    print("Environment:", environment)
    print("Model:", agent.model)
    print("-" * 40)

print("Simulation Finished")
print("Rooms cleaned:", cleaned)
print("Total movements:", movements)
```
