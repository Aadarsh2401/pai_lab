```python
class SmartThermostatAgent:
    def __init__(self):
        self.temperature_levels = ["Low", "Comfortable", "High"]

    def act(self, temperature):
        if temperature == "Low":
            return "Heat"
        elif temperature == "High":
            return "Cool"
        else:
            return "Idle"

    def update_temperature(self, temperature, action):
        index = self.temperature_levels.index(temperature)

        if action == "Heat" and index < 2:
            index += 1
        elif action == "Cool" and index > 0:
            index -= 1

        return self.temperature_levels[index]


# ---- Simulation ----
agent = SmartThermostatAgent()

# Take input from user
initial_temperature = input("Enter initial temperature (Low / Comfortable / High): ").capitalize()

if initial_temperature not in ["Low", "Comfortable", "High"]:
    print("Invalid temperature state!")
else:
    temperature = initial_temperature
    steps = int(input("Enter number of time steps: "))

    print("\nSimulation Started")
    for step in range(steps):
        action = agent.act(temperature)
        print(f"Step {step + 1}: Temperature = {temperature}, Action = {action}")
        temperature = agent.update_temperature(temperature, action)
```

```python
class ElevatorAgent:
    def __init__(self, start_floor=1):
        self.current_floor = start_floor

    def act(self, request_floor):
        if self.current_floor < request_floor:
            return "Move Up"
        elif self.current_floor > request_floor:
            return "Move Down"
        else:
            return "Stay"

    def update_floor(self, action):
        if action == "Move Up" and self.current_floor < 3:
            self.current_floor += 1
        elif action == "Move Down" and self.current_floor > 1:
            self.current_floor -= 1


def simulate_elevator(start_floor, request_floor):
    agent = ElevatorAgent(start_floor)

    print(f"\nElevator starting at Floor {start_floor}")
    print(f"Request received for Floor {request_floor}")

    step = 1
    while True:
        action = agent.act(request_floor)
        print(f"Step {step}: Current Floor = {agent.current_floor}, Action = {action}")

        if action == "Stay":
            print("Goal achieved: Elevator has reached the requested floor.")
            break

        agent.update_floor(action)
        step += 1


# ---- Simulations ----
simulate_elevator(1, 3)
simulate_elevator(3, 1)
simulate_elevator(2, 2)
```

```python
class TrafficSignalAgent:
    def __init__(self):
        self.cost = {
            "Low": 1,
            "High": 5
        }

    def calculate_utility(self, road_a, road_b, action):
        if action == "Green_A":
            total_wait = self.cost[road_b]
        else:  # Green_B
            total_wait = self.cost[road_a]

        return -total_wait  # negative waiting cost

    def choose_action(self, road_a, road_b):
        utility_a = self.calculate_utility(road_a, road_b, "Green_A")
        utility_b = self.calculate_utility(road_a, road_b, "Green_B")

        if utility_a > utility_b:
            return "Green for Road A", utility_a
        else:
            return "Green for Road B", utility_b


def simulate(road_a, road_b):
    agent = TrafficSignalAgent()
    action, utility = agent.choose_action(road_a, road_b)

    print(f"Traffic 상태 → Road A: {road_a}, Road B: {road_b}")
    print(f"Chosen Action: {action}")
    print(f"Utility Value: {utility}\n")


# ---- Simulations ----
simulate("High", "Low")
simulate("Low", "High")
simulate("High", "High")
simulate("Low", "Low")
```

```python
class WarehouseRobotAgent:
    def __init__(self, grid, start, target):
        self.grid = grid
        self.position = start
        self.target = target
        self.visited = set()
        self.rows = len(grid)
        self.cols = len(grid[0])

    def is_valid(self, r, c):
        return (0 <= r < self.rows and
                0 <= c < self.cols and
                self.grid[r][c] == 0 and
                (r, c) not in self.visited)

    def move(self):
        r, c = self.position
        self.visited.add((r, c))

        moves = {
            "Up": (r - 1, c),
            "Down": (r + 1, c),
            "Left": (r, c - 1),
            "Right": (r, c + 1)
        }

        for action, (nr, nc) in moves.items():
            if self.is_valid(nr, nc):
                self.position = (nr, nc)
                print(f"Move {action} → New Position: {self.position}")
                return True
        return False

    def run(self):
        print(f"Start Position: {self.position}")
        print(f"Target Position: {self.target}\n")

        while self.position != self.target:
            if not self.move():
                print("No valid moves left. Navigation failed.")
                return

        print("\nTarget reached!")
        print("Action: Pick Item")


# 0 = Free cell, 1 = Obstacle
warehouse_grid = [
    [0, 0, 1, 0],
    [1, 0, 1, 0],
    [0, 0, 0, 0],
    [0, 1, 1, 0]
]

start = (0, 0)
target = (3, 3)

agent = WarehouseRobotAgent(warehouse_grid, start, target)
agent.run()
```

```python
import random

class RecommendationAgent:
    def __init__(self):
        self.scores = {"A": 0, "B": 0}
        self.exploration_rate = 0.2  # 20% exploration

    # Performance Element
    def recommend(self):
        if random.random() < self.exploration_rate:
            return random.choice(["A", "B"])  # Problem Generator
        return "A" if self.scores["A"] >= self.scores["B"] else "B"

    # Learning Element
    def learn(self, product, feedback):
        if feedback == "Like":
            self.scores[product] += 1
        else:
            self.scores[product] -= 1


# Simulated user behavior (Critic)
def user_feedback(product):
    preferences = {"A": 0.7, "B": 0.3}  # User prefers A
    return "Like" if random.random() < preferences[product] else "Dislike"


# ---- Simulation ----
agent = RecommendationAgent()

print("Online Recommendation Simulation\n")

for session in range(1, 11):
    product = agent.recommend()
    feedback = user_feedback(product)
    agent.learn(product, feedback)

    print(f"Session {session}: Recommended Product {product}, Feedback = {feedback}")
    print(f"Scores → A: {agent.scores['A']}, B: {agent.scores['B']}\n")
```
