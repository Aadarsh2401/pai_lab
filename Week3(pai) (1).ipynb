```python
from collections import deque

def bfs(graph, start, goal):
    queue = deque()
    queue.append([start])

    visited = set()

    nodes_expanded = 0

    while queue:
        path = queue.popleft()
        current_node = path[-1]

        # Goal test
        if current_node == goal:
            return path, nodes_expanded

        #  node if not visited
        if current_node not in visited:
            visited.add(current_node)
            nodes_expanded += 1

            #  successors
            for neighbor in graph[current_node]:
                new_path = path + [neighbor]
                queue.append(new_path)

    # If goal not reachable
    return None, nodes_expanded
```

```python
maze = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E', 'G'],
    'G': []
}
```

```python
start_node = 'A'
goal_node = 'G'

path, expanded = bfs(maze, start_node, goal_node)

print("Path found:", path)
print("Nodes expanded:", expanded)
```

```python
def dfs(graph, current, goal, path):
    path.append(current)

    # Goal
    if current == goal:
        return path

    for neighbor in graph[current]:
        if neighbor not in path:  # Avoid cycles
            result = dfs(graph, neighbor, goal, path.copy())
            if result is not None:
                return result

    return None
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': ['G'],
    'G': []
}

start_node = 'A'
goal_node = 'G'

path = dfs(graph, start_node, goal_node, [])

print("Path found using DFS:", path)
```

```python
from collections import deque

def get_neighbors(state):
    neighbors = []
    index = state.index(0)
    row, col = index // 3, index % 3

    moves = {
        'UP': (-1, 0),
        'DOWN': (1, 0),
        'LEFT': (0, -1),
        'RIGHT': (0, 1)
    }

    for move in moves.values():
        new_row, new_col = row + move[0], col + move[1]
        if 0 <= new_row < 3 and 0 <= new_col < 3:
            new_index = new_row * 3 + new_col
            new_state = list(state)
            new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
            neighbors.append(tuple(new_state))

    return neighbors


def bfs_8_puzzle(start, goal):
    queue = deque()
    queue.append([start])

    visited = set()
    visited.add(start)

    states_expanded = 0

    while queue:
        path = queue.popleft()
        current = path[-1]

        if current == goal:
            return path, states_expanded

        states_expanded += 1

        for neighbor in get_neighbors(current):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path + [neighbor])

    return None, states_expanded
```

```python
start_state = (1, 2, 3,
               4, 0, 5,
               6, 7, 8)

goal_state = (1, 2, 3,
              4, 5, 0,
              6, 7, 8)

solution, expanded = bfs_8_puzzle(start_state, goal_state)

print("Solution path:")
for step in solution:
    print(step)

print("States expanded:", expanded)
```

```python
from collections import deque
import time

def bfs_metrics(graph, start, goal):
    start_time = time.time()

    queue = deque([[start]])
    visited = set()
    nodes_expanded = 0
    max_queue_size = 1

    while queue:
        max_queue_size = max(max_queue_size, len(queue))
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            return path, nodes_expanded, time.time() - start_time, max_queue_size

        if node not in visited:
            visited.add(node)
            nodes_expanded += 1
            for neighbor in graph[node]:
                queue.append(path + [neighbor])

    return None, nodes_expanded, time.time() - start_time, max_queue_size
```

```python
def dfs_metrics(graph, node, goal, visited, depth, stats):
    stats['max_depth'] = max(stats['max_depth'], depth)
    stats['nodes_expanded'] += 1

    if node == goal:
        return [node]

    for neighbor in graph[node]:
        if neighbor not in visited:
            visited.add(neighbor)
            result = dfs_metrics(graph, neighbor, goal, visited, depth + 1, stats)
            if result:
                return [node] + result

    return None
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': ['G'],
    'G': []
}

start, goal = 'A', 'G'

# BFS
bfs_result = bfs_metrics(graph, start, goal)

# DFS
import time
dfs_stats = {'nodes_expanded': 0, 'max_depth': 0}
dfs_start = time.time()
dfs_path = dfs_metrics(graph, start, goal, {start}, 0, dfs_stats)
dfs_time = time.time() - dfs_start

print("\nBFS Results:")
print("Path:", bfs_result[0])
print("Nodes Expanded:", bfs_result[1])
print("Time:", bfs_result[2])
print("Max Queue Size:", bfs_result[3])

print("\nDFS Results:")
print("Path:", dfs_path)
print("Nodes Expanded:", dfs_stats['nodes_expanded'])
print("Time:", dfs_time)
print("Max Recursion Depth:", dfs_stats['max_depth'])
```
